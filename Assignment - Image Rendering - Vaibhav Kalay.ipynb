### **ANSWER FOR Q-3 --- *equalize_adapthist* **

**Syntax:**<br>*skimage.exposure.equalize_adapthist(image, kernel_size=None, clip_limit=0.01, nbins=256)* 
<br>This command is used for local contrast enhancement, that uses histograms calculated over different regions of the image. <br>Local details can therefore be enhanced even in regions that are darker or lighter than most of the image. Its Parameters are as below:

**Parameters:**
* image : image : (M, N[, C]) <br>ndarray
Input image.
* kernel_size: integer or list-like, optional
<br>Defines the shape of contextual regions used in the algorithm. 
* clip_limit : float, optional
<br>Clipping limit, normalized between 0 and 1.
* nbins : int, optional
Number of gray bins for histogram (“data range”).

**Returns:**
<br>out : (M, N[, C]) ndarray
<br>Equalized image.

**Brief Explanation:**
<br>
Adaptive histogram equalization (AHE) is an image processing technique used to improve contrast in images by decimating the image into small rectangular areas, and enhances the contrast of these areas by adjusting their local histograms. The difference compared to ordinary histogram equalization is that the adaptive method computes several histograms, each corresponding to a distinct section of the image, and uses them to redistribute the pixel values of the image. It is therefore suitable for improving the local contrast and enhancing the definitions of edges in each region of an image. However, the limitation of this method is that it has a tendency to overamplify noise in relatively homogeneous regions of an image. Advancement of it also called contrast limited adaptive histogram equalization (CLAHE) prevents this by limiting the amplification.


### **ANSWER FOR Q-4 --- *equalize_hist vs. equalize_adapthist* ** 

On performing **equlize_hist** on the Image, we observe that almost all the pixels are getting utilized to generate an enhanced image and there is no scope of further enhancement of user specific information. This method is a global method wherin performing any changes to view an area, it shall affect the entire image as well as its corresponding histogram. <br>Whereas in **equalize_adapthist**, we find that we have been provided with an option to vary the contrast of some specific portion which helps us to adjust the contrast as per our requirements. We have that widget to toggle and change the contrast of the image as per our needs and requirements which is a very handy and efficient tool. However, it does not affect the entire image and only area specific information is highlighted.
<br><br>We can thus conclude that the adaptive histogram equalization is better compared to global equilization method. It has user friendly interface that allows to toggle pixels to vary the contrast and enhance the rendered image in order to make a certain region of study more specific which is not possible with the global method. 
The only disadvantage of adaptive method is that if noise is present in the dark or light portions, it too gets enhanced while enhancing the neighborhood areas and it affects the entire image quality considerably. Since global histogram equalization performs enhancement on the entire image, there is no huge impact on the final output even if noise is enhanced.
