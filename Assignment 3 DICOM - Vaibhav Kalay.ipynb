**4F** 

1. (0028, 0030) Pixel Spacing                       DS: ['0.732421875', '0.732421875']<br>
Physical distance between the center of each pixel, specified by a numeric pair.
2. (0020, 0037) Image Orientation (Patient)         DS: ['1', '0', '0', '0', '1', '0']<br>
The direction of the first row and the first column with respect to the patient.
3. (0020, 0032) Image Position (Patient)            DS: ['-180.72742462158', '-227.5', '64.1605377197265']<br>
The x, y, and z coordinates of the center of the first voxel of the image.
4. (0018, 0088) Spacing Between Slices              DS: '8'<br>
Spacing between slices, in mm. The spacing is measured from the center-to-center of each slice.
5. (0020, 1041) Slice Location                      DS: '112.000003814697'<br>
The relative position of the image plane expressed as relative to a specific reference point.

Pixel Spacing represents the distance between two consecutive points. Image Orientation and Position represents the co-ordinates of the image. Slice is used to seperate the DICOM Image into finite and parallel segements concatenating which would result into an entire image formation. 
These attributes are extremely useful to describe the geometry of the patient.
<br> Thus we can use the above attributes and can define the geometry of the patient's image. These attributes give detailed and entire information so as to recreate a 3D image.

Digital Imaging and Communications in Medicine (DICOM) Standard specifies data interchange protocol, digital image format, and file structure for biomedical images and image-related information. The fundamental concepts of the DICOM message protocol, services, and information objects are reviewed as background for a detailed discussion of the functionality of DICOM.<br>
PACS are medical information systems (technically consisting of associated hardware and software) designed and used to process digital medical imaging. PACS comprises of digital image acquisition devices (modalities–such  as  computed  tomography  (CT)  scanners, X-rays or 
ultrasound etc), archives for digital image, and  workstations. DICOM packages include information regarding patient information, image acquisition, and the list of images that correspond to axial sections to form 3D images.<br>
DICOM addresses five areas: (1) network image management, (2) network image interpretation management, (3) network print management, (4) imaging procedure management, (5) off-line storage media management.
